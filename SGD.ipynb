```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.show()
```

```python
x_data = np.array(x)
y_data = np.array(y)
```

```python
a = 0 # 초기 기울기 0으로 초기화
b = 0 # 초기 절편 0으로 초기화

# 학습률
lr = 0.03

# 반복 횟수
epochs = 2001

for i in range(epochs):
    y_pred = a*x_data + b    # y_hat
    error = y_data - y_pred    # Error
    
    # MSE를 a에 대해 편미분한 값
    a_diff = -(2/len(x_data)) * sum(error * x_data)
    
    # MSE를 b에 대해 편미분한 값
    b_diff = -(2/len(x_data)) * sum(error)
    
    a -= lr * a_diff
    b -= lr * b_diff
    
    if i % 100 == 0:
        print("a_diff = {}".format(a_diff))
        print("b_diff = {}".format(b_diff))
        print("epoch = {}, a = {:.4f}, b = {:.4f}".format(i, a, b))
        print()
    
```

```python
y_pred = a*x_data + b
plt.scatter(x ,y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```

# 다중 선형 회귀

```python
data = [[2, 0, 81], [4, 4,93], [6, 2, 91], [8, 3, 97]]
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y  = [i[2] for i in data]
```

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection = '3d')
ax.set_xlabel("study_hours")
ax.set_ylabel("private_class")
ax.set_zlabel("score")
ax.scatter(x1, x2, y)
plt.show()
```

```python
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data  = np.array(y)
```

```python
a1 = 0
a2 = 0
b  = 0

lr = 0.025
epochs = 2001

for i in range(epochs):
    y_pred = a1*x1_data + a2*x2_data + b
    error = y_data - y_pred
    
    a1_diff = -(2/len(x1_data)) * sum(error * x1_data)
    a2_diff = -(2/len(x2_data)) * sum(error * x2_data)
    b_diff = -(2/len(x1_data)) * sum(error)
    
    a1 -= lr * a1_diff
    a2 -= lr * a2_diff
    b -= lr * b_diff
    
    if i % 100 == 0:
        print('epoch = {}, a1 = {:.4f}, a2 = {:.4f}, b = {:.4f}'.format(i, a1, a2, b))
```
